```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create a scatter plot specifying "o"
plt.plot(x_axis, y_axis, 'o')
```

```python
# Create a scatter plot
plt.scatter(x_axis, y_axis)
```

```python
# Create the plot.
plt.scatter(y_axis, x_axis, color="red", linewidth=2, label='Chicago')
plt.gca().invert_yaxis()
# Create labels for the x and y axes.
plt.xlabel("Date")
plt.ylabel("Fare($)")
# Create a title.
plt.title("PyBer Fare by Month")
# Add the legend.
plt.legend()
# Change figue size to 20 pixels by 10 pixels high.
plt.figure(figsize=(20,10))
```

```python
# Create a bubble chart.
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Make bigger a bubble chart and multiply the data in the y-axis by any value.
y_axis_larger = []
for data in y_axis:
  y_axis_larger.append(data*3)

plt.scatter(x_axis, y_axis, s=y_axis_larger)
```

```python
# Applying list comprehension technique
plt.scatter(x_axis, y_axis, s = [i *3 for i in y_axis])
```

```python
# Create a Scatter Plot Using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
# Create a Bubble Chart Using the Object-Oriented Interface.
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, s=y_axis)
```

```python
# Skill Drill
fig, ax = plt.subplots()
ax.scatter(y_axis, x_axis, s = [i *5 for i in y_axis], color="skyblue", edgecolor="k", linewidth=2, alpha=.8, label='Boston')
ax.set(xlabel="Date", ylabel="Fare($)", title="PyBer Fare by Month")
ax.set_xlim(0,50)
ax.invert_yaxis ()
ax.legend()
```
